```python
import pandas as  pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
#load dataset
trizah=pd.read_csv('/accident_data.csv')

le = LabelEncoder()
trizah['Weather'] = le.fit_transform(trizah['Weather'])

#extract the independent variable and dependent variable
x= trizah['Weather'].values.reshape(-1, 1)
y = trizah['Severity'].values.reshape(-1, 1)

#create and fit a linear regression model
regressor =LinearRegression()
regressor.fit(x, y)

#plot datapoints
trizah.plot(x='Weather', y='Severity', style='o', label='Datapoints')
#Overlay the rgression line
plt.plot(x, regressor.predict(x), color='red', linewidth=2, label='Regression Line')

plt.xlabel('Weather')
plt.ylabel('Severity')
plt.legend()
plt.show()
```
